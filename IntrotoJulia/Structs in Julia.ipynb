```julia
mutable struct Person
    name::String
    age::Float64
    isActive

    function Person(name, age)
        new(name, age, true)
    end
end
```

```julia
newPerson1 = Person("Amit", 28)
```

```julia
function birthday(person::Person)
    person.age += 1
end
```

```julia
birthday(newPerson1)
```

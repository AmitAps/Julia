# conditions

if *condition 1*

    *option 1*
    
elseif *condition 2*

    *option 2*
    
else 

    *option 3*
    
end


```julia
x = 3
y = 5
```

```julia
if x > y 
    println("$x is larger than $y")
elseif y > x
    println("$y is larger than $x")
else
    println("$x and $y are equal")
end
```

```julia
if x > y
    x
else 
    y
end
```

# ternary operator with the syntax
`a ? b : c`

which equates to 

## if a
     b
## else
     c
## end

```julia
(x > y) ? x : y 
```

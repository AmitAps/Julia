# Loops

1. `while` loops
2. `for` loops

# While loops
The syntax for a `while` is

while *condition*

    *loop body*  
    
end

```julia
n = 0
while n < 10
    n += 1
    println(n)
end
```

```julia
myfriends = ["Aps", "As", "Aakshi", "Aditya","Mk"]
startindex = 1
while startindex <= length(myfriends)
    friend = myfriends[startindex]
    println("Hi $friend , it's great to see you again!")
    startindex += 1
end
```

# for loops

for *var* in *loop iterable*
    *loop body*
end

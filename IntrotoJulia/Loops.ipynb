# While loops
The syntax for a `while` is

while *condition*

    *loop body*  
    
end

```julia
n = 0
while n < 10
    n += 1
    println(n)
end
```

```julia
myfriends = ["Aps", "As", "Aakshi", "Aditya","Mk"]
startindex = 1
while startindex <= length(myfriends)
    friend = myfriends[startindex]
    println("Hi $friend , it's great to see you again!")
    startindex += 1
end
```

# for loops

for *var* in *loop iterable*

    *loop body*
    
end

```julia
for n in 1:10
    println(n)
end
```

```julia
myfriends = ["Mk", "Aps","Singh", "Sakshi", "FK"]
for friend in myfriends
    println("Hi My friend $friend ., It's great to see you again!")
end
```

```julia
for n = 1:10
    println(n)
end
```

```julia
for n ∈ 1:10
    println(n)
end
```

```julia
m, n = 5, 5
A = zeros(m, n)
```

```julia
for i in 1:m
    for j in 1:n
        A[i, j] = i+j
    end
end
A
```

```julia
B = zeros(m, n)
```

```julia
for i in 1:m, j in 1:n
    B[i, j] = i+j
end
B
```
